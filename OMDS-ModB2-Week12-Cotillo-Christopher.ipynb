# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

A: The first plot shows side by side comparisons of the series values listed below. For presentation, this can be confusing to a leader who is viewing this for the first time. It is key to make presentations easy to navigate through simple figures. The second figure is favorable over the first for presentations. One of the obvious pros of the first plot is the zoomed in effect that it portrays, unlike the second plot that requires a larger overall plot. 

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 187-205 of the Storytelling With Data book as best you can. (Chapter Eight). You do not have to get the exact data values right, just the overall look and feel.

```python

data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [120, 135, 150, 160, 170, 180]
}
df = pd.DataFrame(data)
plt.figure(figsize=(8, 4))
plt.plot(df['Month'], df['Sales'], marker='o', color='steelblue', linewidth=2)
plt.title('Monthly Sales', fontsize=14, weight='bold')
plt.grid(False)
plt.xlabel('Month')
plt.ylabel('Sales')
```
